# Texas Public School Discipline Rankings 2017-2018

This report creates rankings for all Texas public schools by discipline types. 
* Rankings are based on the percentage of discipline actions compared to campus cumulative year end enrollment.  
* A school with the lowest rank has the highest percentage of a particular discipline type in the State of Texas. 
* Data used in this report comes from: https://rptsvr1.tea.texas.gov/adhocrpt/Disciplinary_Data_Products/Download_All_Campuses.html
* There were 8,612 campuses considered in this study.
* There were 1200 school districts considered in this study.

## Region Specific Discipline Rankings 2017-2018
When considering public schools in Region 11, Hebron High School discipline rankings move even higher.

**Findings are as follows:**
* There are 904 public schools considered within Region 11.
* Hebron ranks 95th out of 904 public schools for discipline records as a percentage of Campus Cumulative Year End Enrollment.  
* Hebron High School's total campus discipline records percentage rank falls in the top 13.7% of all Region 11 schools.
* Hebron High School's discipline population percentage rank falls in the top 23.2% of all Region 11 schools.
* Hebron High School's IIS percentage rank falls in the top 22.2% of all Region 11 schools.

```python
#Isolate Region 11
regDisc2018 = pd.DataFrame(disciplinePct2018[(disciplinePct2018['REGION'] == 11)]) 

regDisc2018.reset_index(inplace=True)
regDisc2018.drop(['index'], axis=1, inplace=True)
print('Total schools considered in Region 11: ' + str(len(regDisc2018.index))) 
```

```python
#Create discipline rankings only within Region 11 
regDisc2018['DisciplinePopPct_Region_Rank'] = \
    regDisc2018['DisciplinePopPct'].rank(ascending=False,pct=True)
                                                   
regDisc2018['DisciplineRecordCtPct_Region_Rank'] = \
    regDisc2018['DisciplineRecordCtPct'].rank(ascending=False,pct=True)

regDisc2018['ISS_Pct_Region_Rank'] = \
    regDisc2018['ISS_Pct'].rank(ascending=False,pct=True)  

    
regDisc2018.drop(['JJAEP_Pct','DAEP_Pct','Discr_DAEP_Pct','ExpelledPct',
                  'Discr_Expul_Pct', 'Discr_Expul_JJAEP_Pct','Mand_DAEP_Pct',
                  'Mand_Expul_Pct','Mand_Expul_JJAEP_Pct', 'Suspension_Pct']
                  , axis=1, inplace=True)

regDisc2018.sort_values(by=['DisciplineRecordCtPct_Region_Rank'], inplace=True)
regDisc2018.reset_index(inplace=True)
regDisc2018.drop(['index'], axis=1, inplace=True)
regDisc2018.head(100)
```
